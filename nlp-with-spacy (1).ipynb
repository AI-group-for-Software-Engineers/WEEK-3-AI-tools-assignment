```python
# Path to the training file
file_path = '/kaggle/input/amazonreviews/train.ft.txt.bz2'

# Read lines from bz2 file
import bz2

reviews = []
labels = []

with bz2.open(file_path, 'rt', encoding='utf-8') as f:
    for line in f:
        line = line.strip()
        if line:
            # Split label and text
            parts = line.split(' ', 1)  # split at first space
            label = parts[0]            # e.g., '__label__1' or '__label__2'
            review_text = parts[1] if len(parts) > 1 else ''
            labels.append(label)
            reviews.append(review_text)

# Quick check
print(f"Total reviews loaded: {len(reviews)}")
print("Sample review:", reviews[0])
print("Sample label:", labels[0])
```

```python
import spacy

# Load English small model
nlp = spacy.load("en_core_web_sm")
```

```python
# Extract entities from first 10 reviews (for demo)
for i, review in enumerate(reviews[:10]):
    doc = nlp(review)
    entities = [(ent.text, ent.label_) for ent in doc.ents if ent.label_ in ['ORG', 'PRODUCT']]
    print(f"Review {i+1}: {review}")
    print(f"Extracted Entities: {entities}")
    print('-'*50)
```

```python
# Define simple keywords for sentiment
positive_words = ["good", "great", "excellent", "love", "awesome", "best", "beautiful", "impress"]
negative_words = ["bad", "terrible", "poor", "hate", "worst", "disappointed", "crude", "boring"]

def sentiment_analysis(text):
    text = text.lower()
    pos_count = sum([text.count(word) for word in positive_words])
    neg_count = sum([text.count(word) for word in negative_words])
    
    if pos_count > neg_count:
        return "Positive"
    elif neg_count > pos_count:
        return "Negative"
    else:
        return "Neutral"

# Apply to first 150 reviews
for i, review in enumerate(reviews[:150]):
    sentiment = sentiment_analysis(review)
    print(f"Review {i+1} Sentiment: {sentiment}")
```
